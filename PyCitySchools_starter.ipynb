```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
total_schools = school_data["School ID"].count()
total_students = student_data["Student ID"].count()
total_budget = school_data["budget"].sum()
avg_math = student_data["math_score"].mean()
avg_reading = student_data["reading_score"].mean()
overall_avg_scores_rate = (avg_math + avg_reading)/2
```

```python
# use comparison work with df list to find students group whose passed test score
mathPass70 =  student_data["math_score"] >= 70    
mathPass_students = len(student_data.loc[mathPass70])
readPass70 = student_data["reading_score"] >= 70
readPass_students = len(student_data.loc[readPass70])
```

```python
# Calculate the percentage of students with a passing score (70 or greater)
mathPass = (mathPass_students / total_students)*100
readPass = (readPass_students / total_students)*100
```

```python
# Create a dataframe to hold the above results
district_summary_df = pd.DataFrame({"Total Schools":[total_schools],
                                   "Total Students":[total_students],
                                   "Total Budget":[total_budget],
                                   "Average Math Score":[avg_math],
                                   "Average Reading Score":[avg_reading],
                                   "Overall Average Score":[overall_avg_scores_rate],
                                   "% Passing Math":[mathPass],
                                   "% Passing Reading":[readPass]})
```

```python
# Optional: give the displayed data cleaner formatting by 
# converting district_summary_df (a dictionary) to a list of lists, District_Summary, so that
# the value of list can be diaplayed to cleaner format
District_Summary = district_summary_df[["Total Schools",
                                       "Total Students",
                                       "Total Budget",
                                       "Average Math Score",
                                       "Average Reading Score",
                                       "Overall Average Score",
                                       "% Passing Math",
                                       "% Passing Reading"]]
                                       
District_Summary["Total Students"] = district_summary_df["Total Students"].map("{:,}".format)
District_Summary["Total Budget"] = district_summary_df["Total Budget"].map("${:.2f}".format)
District_Summary["Average Math Score"] = district_summary_df["Average Math Score"].map("{:.2f}".format)
District_Summary["Average Reading Score"] = district_summary_df["Average Reading Score"].map("{:.2f}".format)
District_Summary["Overall Average Score"] = district_summary_df["Overall Average Score"].map("{:.2f}".format)
District_Summary["% Passing Math"] = district_summary_df["% Passing Math"].map("{:.2f}%".format)
District_Summary["% Passing Reading"] = district_summary_df["% Passing Reading"].map("{:.2f}%".format)
District_Summary
```

```python
# create a groupby opject for future analysis of each school
grouped_school_data = school_data_complete.groupby(["school_name", "type"])
```

```python
# calculate each reqired value for new DataFrame
grped_total_students = grouped_school_data["Student ID"].count()
grped_total_school_budget = grouped_school_data["budget"].sum()
grped_per_student_budget = (grped_total_school_budget / grped_total_students)
grped_avg_math = grouped_school_data["math_score"].mean()
grped_avg_reading = grouped_school_data["reading_score"].mean()
```

```python
# use comparison work with grouped list to find students group whose passed test score
mathPass70_perSchool = school_data_complete[school_data_complete["math_score"] >= 70]
readPass70_perSchool = school_data_complete[school_data_complete["reading_score"] >= 70]
grped_mathPass70 = mathPass70_perSchool.groupby("school_name")
grped_readPass70 = readPass70_perSchool.groupby("school_name")
grped_mathPass70_students = grped_mathPass70["Student ID"].count()
grped_readPass70_students = grped_readPass70["Student ID"].count()
grped_percent_mathPass70 = (grped_mathPass70_students/grped_total_students)*100
grped_percent_readPass70 = (grped_readPass70_students/grped_total_students)*100
grped_overall_percentPass_rate = (grped_percent_mathPass70 + grped_percent_readPass70)/2
```

```python
# Storing above results into a dictionary DataFrame
grouped_school_data_df = pd.DataFrame({"Total Students": grped_total_students,
                                       "Total School Budget": grped_total_school_budget, 
                                       "Per Student Budget": grped_per_student_budget,
                                       "Average Math Score": grped_avg_math,
                                       "Average Reading Score": grped_avg_reading,   
                                       "% Passing Math": grped_percent_mathPass70,
                                       "% Passing Reading": grped_percent_readPass70,
                                       "% Overall Passing Rate":grped_overall_percentPass_rate})
```

```python
# convert the dictionary DataFrame into a list of lists DataFrame
School_Summary = grouped_school_data_df[["Total Students",
                                         "Total School Budget",
                                         "Per Student Budget",
                                         "Average Math Score",
                                         "Average Reading Score",
                                         "% Passing Math",
                                         "% Passing Reading",
                                         "% Overall Passing Rate"]]

# display the data to cleaner formatting
School_Summary["Total Students"] = grouped_school_data_df["Total Students"].map("{:,}".format)
School_Summary["Total School Budget"] = grouped_school_data_df["Total School Budget"].map("${:,.2f}".format)
School_Summary["Per Student Budget"] = grouped_school_data_df["Per Student Budget"].map("${:,.2f}".format)
School_Summary["Average Math Score"] = grouped_school_data_df["Average Math Score"].map("{:.2f}".format)
School_Summary["Average Reading Score"] = grouped_school_data_df["Average Reading Score"].map("{:.2f}".format)
School_Summary["% Passing Math"] = grouped_school_data_df["% Passing Math"].map("{:.2f}%".format)
School_Summary["% Passing Reading"] = grouped_school_data_df["% Passing Reading"].map("{:.2f}%".format)
School_Summary["% Overall Passing Rate"] = grouped_school_data_df["% Overall Passing Rate"].map("{:.2f}%".format)
School_Summary
```
